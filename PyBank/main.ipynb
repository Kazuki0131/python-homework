```python
#import the csv and Path module
import csv
from pathlib import Path
```

```python
#set the path of budget_data.csv
csvpath = Path("Resources/budget_data.csv")

#create and intialize variables
##################
bank_data = []
count = 0
total_month = 0
#total of profits
total = 0
previous_profit = 0

#number of changes of profits
num_changes = 0
#total of changes of profits
sum_changes = 0

greatest_increase = 0
date_of_greatest_increase = ""
greatest_decrease = 0
date_of_greatest_decrease = ""
###################

#open the budget_data.csv, read the data, and calculate each result
with open(csvpath) as f:
    csv_data = csv.reader(f, delimiter=',')
    
    #skip the header of budget data
    next(csv_data)
    
    #iterate over rows of csv data
    for line in csv_data:
        
        if count == 0:
            #convert the profits data to integer (line[1] contains the profits data)
            line[1] = int(line[1])
            
            #store the firt profits data into previous_profits variable
            #skip calculating the change of profits
            previous_profit = int(line[1])
            
        else:
            #convert the profits data to integer (line[1] contains the profits data)
            line[1] = int(line[1])
            
            #calculate the change between previous and current profits)
            change = line[1] - previous_profit
            line.append(change)
            
            #identify the greatest increase and decrease of change of profits
            if num_changes == 0:
                #store the change of proifts in the greatest_increase and greatet_decrease variables if the change of profits is the first one
                greatest_increase = change
                date_of_greatest_increase = line[0]
                greatest_decrease = change
                date_of_greatest_decrease = line[0]
            elif change > greatest_increase:
                greatest_increase = change
                date_of_greatest_increase = line[0]
            elif change < greatest_decrease:
                greatest_decrease = change
                date_of_greatest_decrease = line[0]
                
            #calculate the number of changes and total of changes
            num_changes += 1
            sum_changes += change
            
            #store the firt profits data into previous_profits variable
            previous_profit = int(line[1])
        
        #increase the count
        count += 1
        
        #sum up the profits
        total += line[1]
        
        #add the line data into bank_data list
        bank_data.append(line)
```

```python
#calculate the total months of bank data
total_months = len(bank_data)
#calculate the average change of profits
average_change = round(sum_changes/num_changes,2)
```

```python
#print out the results
analysis = "Financial Analysis\n"
analysis = analysis + "------------------------------\n"
analysis = analysis + "Total Months: " + str(total_months) + "\n"
analysis = analysis + "Total: $" + str(total) + "\n"
analysis = analysis + "Average Cahnge: $" + str(average_change) + "\n"
analysis = analysis + "Greatest Increase in Profits: " + date_of_greatest_increase + " ($" + str(greatest_increase) + ")\n"
analysis = analysis + "Greatest Decrease in Profits: " + date_of_greatest_decrease + " ($" + str(greatest_decrease) + ")\n"

print(analysis)
```

```python
#set a path for financial_analysis.txt to store the result
output_path = Path("financial_analysis.txt")
```

```python
#open the financial_analysis.txt and write the reulst in the file
with open(output_path, 'w') as file:
    file.write(analysis)
```
